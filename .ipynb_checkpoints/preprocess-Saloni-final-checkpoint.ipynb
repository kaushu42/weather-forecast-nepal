```python
import os
import pandas as pd
```

```python
main_df = pd.DataFrame()
```

```python
xls = pd.ExcelFile('./data/2016/Jan 2016.xlsx')
xls.sheet_names
```

# For visibility data

```python
df = pd.read_excel(xls, sheet_name='visibility', skiprows=7, usecols="B:H")
df.head()
```

```python
df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
df.columns = ['Time', 'visibility']
df = pd.get_dummies(df)
df.head()
```

```python
main_df = pd.concat([main_df, df])
main_df
```

# For dew point

```python
df = pd.read_excel(xls, sheet_name='Dew pt', skiprows=7, usecols="C:H")
df.head()
```

```python
df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
df.columns = ['Time', 'dew_pt']
df = pd.get_dummies(df)
df.head()
```

```python
main_df['dew_pt'] = df['dew_pt']
main_df.head()
```

# For QFE

```python
df = pd.read_excel(xls, sheet_name='QFE', skiprows=10, usecols="C:H")
df.head()
```

```python
df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
df.columns = ['Time', 'QFE']
df = pd.get_dummies(df)
df.head()
```

```python
main_df['QFE'] = df['QFE']
main_df.head()
```

# For Cloud

```python
df = pd.read_excel(xls, sheet_name='cloud', skiprows=3, header=[0, 1])
df.head()
```

```python
df = df.set_index(('Day/Time(UTC)', 'Unnamed: 0_level_1')).stack(level=0).reset_index()
df = df.set_index(('Day/Time(UTC)', 'Unnamed: 0_level_1'))
df.index.name = 'Day/Time(UTC)'
df = pd.get_dummies(df)
```

```python
main_df['LC'] = df['LC']
main_df['TC'] = df['TC']
main_df.head()
```

# For the Temperature

```python
df = pd.read_excel(xls, sheet_name='Temperature', skiprows=3, header=[0, 1]).iloc[:, 0:16]
df.drop(('Unnamed: 0_level_0', 'Unnamed: 0_level_1'), axis=1, inplace=True)
```

```python
df = df.set_index(('Day', 'Unnamed: 1_level_1')).stack(level=0).reset_index()
df = df.set_index(('Day', 'Unnamed: 1_level_1'))
df.index.name = 'Day/Time(UTC)'
df = pd.get_dummies(df)
```

```python
main_df['DRY'] = df['DRY']
main_df['RH'] = df['RH']
main_df['WET'] = df['WET']
main_df.head()
```

# Iteration over files

```python
import os
import pandas as pd
```

```python
filenamelist = []
PATH = './data/'
for year in [2016, 2017, 2018]:
    FILEPATH = f'{PATH}{year}'
    files = os.listdir(FILEPATH)
    for file in files:
        try:
            if file.endswith(".xlsx"):
                main_df = pd.DataFrame()
                filenamelist.append(file)
                filepath = subdir + os.sep + file
                xls = pd.ExcelFile(f'{FILEPATH}/{file}')
                df = pd.read_excel(xls, sheet_name='visibility', skiprows=7, usecol="B:G")

                df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
                df.columns = ['Time', 'visibility']
                df = pd.get_dummies(df)
                
                main_df = pd.concat([main_df, df])
                df = pd.read_excel(xls, sheet_name='Dew pt', skiprows=7, usecols="C:H")
                df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
                df.columns = ['Time', 'dew_pt']
                df = pd.get_dummies(df)
                main_df['dew_pt'] = df['dew_pt']
                df = pd.read_excel(xls, sheet_name='QFE', skiprows=10, usecols="C:H")
                df = df.set_index('Day/Time(UTC)').stack().reset_index().set_index('Day/Time(UTC)')
                df.columns = ['Time', 'QFE']
                df = pd.get_dummies(df)
                main_df['QFE'] = df['QFE']
                df = pd.read_excel(xls, sheet_name='cloud', skiprows=3, header=[0, 1])
                df = df.set_index(('Day/Time(UTC)', 'Unnamed: 0_level_1')).stack(level=0).reset_index()
                df = df.set_index(('Day/Time(UTC)', 'Unnamed: 0_level_1'))
                df.index.name = 'Day/Time(UTC)'
                df = pd.get_dummies(df)
                main_df['LC'] = df['LC']
                main_df['TC'] = df['TC']
                df = pd.read_excel(xls, sheet_name='Temperature', skiprows=3, header=[0, 1]).iloc[:, 0:16]
                df.drop(('Unnamed: 0_level_0', 'Unnamed: 0_level_1'), axis=1, inplace=True)
                df = df.set_index(('Day', 'Unnamed: 1_level_1')).stack(level=0).reset_index()
                df = df.set_index(('Day', 'Unnamed: 1_level_1'))
                df.index.name = 'Day/Time(UTC)'
                df = pd.get_dummies(df)
                main_df['DRY'] = df['DRY']
                main_df['RH'] = df['RH']
                main_df['WET'] = df['WET']
                main_df.to_pickle("./"+ file[:-5] + ".pkl")
        except Exception as e:
            print(file)
```
