```python
import os
import datetime
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def normalize(df):
    normalized_df=(df-df.min())/(df.max()-df.min())
    return normalized_df
```

```python
df = pickle.load(open('./dumps/Jan2016.pkl', 'rb'))
df
```

```python
columns = df.columns
columns
```

```python
sampled = df.resample('D')
```

```python
fig, ax_array = plt.subplots(2, 4, figsize=(18, 8))
for i in range(2):
    for j in range(4):
        title = columns[4*i+j]
        sampled.mean()[title].plot(kind='line', ax=ax_array[i][j], title=title.upper())
```
