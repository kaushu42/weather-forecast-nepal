```python
import os
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
xls = pd.ExcelFile('./data/2016/Jan 2016.xlsx')
main_df = pd.DataFrame()
xls.sheet_names
```

# For the visibility

```python
df = pd.read_excel(xls, sheet_name='visibility', skiprows=3).set_index('Day/Time(UTC)')
df.mean(axis=1).plot(kind='line')
plt.xlabel('Day')
plt.ylabel('Average Visibility')
plt.title('January 2016')
plt.show()
```

# For the dew point

```python
df = pd.read_excel(xls, sheet_name='Dew pt', skiprows=3).set_index('Day/Time(UTC)')
df.mean(axis=1).plot(kind='line')
plt.xlabel('Day')
plt.ylabel('Average Dew Point')
plt.title('January 2016')
plt.show()
```

# QFE

```python
df = pd.read_excel(xls, sheet_name='QFE', skiprows=3).set_index('Day/Time(UTC)')
df.mean(axis=1).plot(kind='line')
plt.xlabel('Day')
plt.ylabel('Average QFE')
plt.title('January 2016')
plt.show()
```

# For cloud

```python
df = pd.read_excel(xls, sheet_name='cloud', skiprows=3, header=[0, 1])
df = df.set_index(('Day/Time(UTC)', 'Unnamed: 0_level_1'))
df = df.stack().mean(axis=1).unstack()
df.plot(kind='line')
plt.xlabel('Day')
plt.ylabel('Average TC/LC')
plt.title('January 2016')
plt.show()
```

# For temperature

```python
df = pd.read_excel(xls, sheet_name='Temperature', skiprows=2, header=[0, 1]).iloc[:, 0:16]
df = df.set_index(('Day', 'Unnamed: 0_level_1'))
df = df.stack().mean(axis=1).unstack()
df.plot(kind='line')
plt.xlabel('Day')
plt.ylabel('Average Humidity')
plt.title('January 2016')
plt.show()
```
